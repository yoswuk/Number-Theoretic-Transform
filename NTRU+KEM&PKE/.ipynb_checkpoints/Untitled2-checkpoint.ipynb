```python
def find_w_order_cyclotomic_trinomial(n,b):

    return 3*n/b;
```

```python
def find_w_cyclotomic_trinomial(n,q,b):

    w_order = find_w_order_cyclotomic_trinomial(n,b);
    
    k = Integer((q-1)/w_order);
    
    ws = [g^k for g in find_generator(q)];
    
    return sorted(list(set(ws)));
```

```python
def gen_tree(n,q,b):
    
    Zq = IntegerModRing(q);
    
    fac = factor(n/b);
    
    if len(fac) == 2:
        Radix2 = fac[0][1];
        Radix3 = fac[1][1];
    
    if len(fac) == 1:
        if fac[0][0] == 2:
            Radix2 = fac[0][1];
            Radix3 = 0;

        if fac[0][0] == 3:
            Radix2 = 0;
            Radix3 = fac[0][1];

    level = Radix2+Radix3;

    tree = zero_matrix(ZZ,level+1,n/b);
    index = zero_matrix(ZZ,n/b,1);
    tree[0,0] = find_w_order_cyclotomic_trinomial(n,b);
    
    #Radix-2 NTT with Cyclotomic Polynomial
    tree[1,0] = tree[0,0] / 6;
    tree[1,1] = 5*tree[0,0] / 6;

    index = [tree[1,0]];

    #Radix-3 NTT
    for i in range(2):
        tree[1,3*i  ] = tree[0,i] / 3;
        tree[1,3*i+1] = tree[0,i] / 3 + tree[0,0]/3;
        tree[1,3*i+2] = tree[0,i] / 3 + 2*tree[0,0]/3;

        index.append(tree[1,3*i  ]);
        index.append(tree[1,3*i  ]*2);
            
    #Radix-2 NTT
    for i in range(6):
        tree[2,2*i  ] = tree[1,i] / 2;
        tree[2,2*i+1] = tree[1,i] / 2 + tree[0,0]/2;

        index.append(tree[2,2*i  ]);

    #Radix-3 NTT
    for i in range(18):
        tree[3,3*i  ] = tree[2  ,  i] / 3;
        tree[3,3*i+1] = tree[2  ,  i] / 3 + tree[0,0]/3;
        tree[3,3*i+2] = tree[2  ,  i] / 3 + 2*tree[0,0]/3;

        index.append(tree[3,3*i  ]);
        index.append(tree[3,3*i  ]*2);

    #Radix-2 NTT
    for l in range(3,level):
    for i in range(36):
        tree[l+1,2*i  ] = tree[l  ,  i] / 2;
        tree[l+1,2*i+1] = tree[l  ,  i] / 2 + tree[0,0]/2;

        index.append(tree[l+1,2*i  ]);
            
            
    return (tree, index);    
```
