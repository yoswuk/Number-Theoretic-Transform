```python
def find_n_cyclotomic_trinomial(lbits,hbits):
    set = [];
    for i in range(1,hbits+1):
        for j in range(0,hbits+1):
            n = 2^i*3^j;
            
            if n > 2^hbits or n < 2^lbits:
                continue;
            else:
                set.append(n);
    
    set.sort();
            
    return set;
```

```python
def find_ntt_prime_cyclotomic(n,b,lbits,hbits):
    
    fac = factor(n);
    
    if len(fac) > 2 or len(fac) == 0:
        #print("does not support cyclotomic trinomial");
        return 0;

    if len(fac) == 1:
        if fac[0][0] != 2:
            #print("does not support cyclotomic trinomial");
            return 0;
            
    if len(fac) == 2:
        if fac[0][0] != 2 or fac[1][0] != 3:
            #print("does not support cyclotomic trinomial");
            return 0;
        
    if ((n % b) != 0) or (n/2 < b):
        #print("please select adequete b");
        return 0;
    
    w_order = 3*n/b;
    
    qs =[];

    k = 1;

    while True:
        q = w_order*k+1;

        k += 1;

        if q < 2^lbits:
            continue;
            
        if q > 2^hbits:
            break;

        if q in Primes():
            qs.append(q);

    return qs;
```

```python
def find_generator(q):

    Zq = IntegerModRing(q);

    gs = range(1,q);

    for x in list(factor(q-1)):
        p = x[0];

        t = [];

        for g in gs:
            if Zq(g)^((q-1)/p) != 1:
                t.append(Zq(g));

        gs = t;

    return gs;
```

```python
def gen_tree(n,q,b,w):
    
    Zq = IntegerModRing(q);
    
    zetas = [];
    
    fac = factor(n/b);
    
    if len(fac) == 2:
        Radix2 = fac[0][1];
        Radix3 = fac[1][1];
    
    if len(fac) == 1:
        if fac[0][0] == 2:
            Radix2 = fac[0][1];
            Radix3 = 0;

        if fac[0][0] == 3:
            Radix2 = 0;
            Radix3 = fac[0][1];

    level = Radix2+Radix3;

    tree = zero_matrix(ZZ,level+1,n/b);
    tree[0,0] = find_w_order_cyclotomic_trinomial(n,b);

    #Radix-2 NTT with Cyclotomic Polynomial
    tree[1,0] = tree[0,0] / 6;
    tree[1,1] = 5*tree[0,0] / 6;

    zetas.append(Zq(w)^tree[1,0]);

    #Radix-3 NTT
    for l in range(1,Radix3+1):
        for i in range(2*3^(l-1)):
            tree[l+1,3*i  ] = tree[l  ,  i] / 3;
            tree[l+1,3*i+1] = tree[l  ,  i] / 3 + tree[0,0]/3;
            tree[l+1,3*i+2] = tree[l  ,  i] / 3 + 2*tree[0,0]/3;

            zetas.append(Zq(w)^tree[l+1,3*i]);
            
    #Radix-2 NTT
    for l in range(Radix3+1,level):
        for i in range(2*3^(Radix3)*2^(l-(Radix3+1))):
            tree[l+1,2*i  ] = tree[l  ,  i] / 2;
            tree[l+1,2*i+1] = tree[l  ,  i] / 2 + tree[0,0]/2;
            
            zetas.append(Zq(w)^tree[l+1,2*i]);
            
    return (zetas, tree);
```

```python
class NTT:
    def __init__(self,n,q,b,w):
        self.n = n;
        self.b = b;
        self.w = w;

        Zq = IntegerModRing(q);

        self.two_inv = Zq(2)^(q-2);
        self.level = Integer(log(n,2));

        self.zetas = [];

        tree = zero_matrix(ZZ,self.level+1,1 << self.level);
        tree[0,0] = self.n;

        for l in range(self.level):
            for i in range(1 << l):
                tree[l+1,2*i  ] = tree[l  ,  i] / 2;
                tree[l+1,2*i+1] = tree[l+1,2*i] + self.n;

                self.zetas.append(Zq(w)^tree[l+1,2*i]);

        print(tree)

    def ntt(self, a):
        k = 0;

        b = a.copy();

        #Radix-2 NTT with Cyclotomic Trinomial
        step = self.n/2;

        zeta = self.zetas[k];
        k += 1;

        for i in range(0,step):
            t = zeta * b[i+step];

            b[i+step] = Zq(b[i]+b[i+step]-t);
            b[i     ] = Zq(b[i]+t);
            
        #Radix-3 NTT            
        while step >= 1:
            
            step /= 3;
            
            for start in range(0,self.n,(step << 1)):
                zeta1 = self.zetas[k]; k += 1;
                zeta2 = self.zetas[k]; k += 1;

                for i in range(start,start+step):       
                    t1 = zeta1 * b[i+step];
                    t2 = zeta2 * b[i+2*step];
                    t3 = t1*self.w;
                    t4 = t2*self.w2; 
                    
                    t1 = t1+t2;
                    t3 = t3+t4;

                    b[i+2*step] = b[i]-(t1+t3);
                    b[i+  step] = b[i]+t3;
                    b[i       ] = b[i]+t1;

        #Radix-2 NTT            
        while step >= 1:
            
            step /= 2;
                        
            for start in range(0,self.n,(step << 1)):
                zeta = self.zetas[k]; k += 1;

                for i in range(start,start+step):
                    t = zeta * b[i+step];

                    b[i+step] = Zq(b[i]-t);
                    b[i     ] = Zq(b[i]+t);

        return b;

    def invntt(self, a):
        k = self.n-2;

        step = 1;

        b = a.copy();

        #Radix-2 InvNTT
        while step <= self.n/2:

            step <<= 1;
            
            for start in range(0,self.n,(step << 1)):
                zeta = self.zetas_inv[k]; k += 1;

                for i in range(start,start+step):
                    t = b[i+step];

                    b[i+step] = Zq(t-b[i]) * zeta;
                    b[i     ] = Zq(t+b[i])
            
        #Radix-3 InvNTT            
        while step >= 1:
            
            step *= 3;
            
            for start in range(0,self.n,(step << 1)):
                zeta1 = self.zetas_inv[k]; k += 1;
                zeta2 = self.zetas_inv[k]; k += 1;

                for i in range(start,start+step):       
                    t1 = b[i+step]*self.w2 + b[i+2*step]*self.w;
                    t2 = b[i+step] + b[i+2*step];
                    
                    b[i+2*step] = zeta2*(r[i] - (t1 + t2));
                    b[i+  step] = zeta1*(r[i] + t1);
                    b[i       ] = b[i] + t2;

        #Radix-2 NTT with Cyclotomic Trinomial
        step = self.n/2;

        zeta = self.zetas[k];
        k += 1;

        for i in range(0,step):
            t = zeta * b[i+step];

            b[i+step] = Zq(b[i]+b[i+step]-t);
            b[i     ] = Zq(b[i]+t);
            
        #Radix-2 InvNTT with Cyclotomic Trinomial
        step <<= 1;
        
        zeta = self.zetas_inv[k];

        for i in range(step):
            t = b[i+step];

            t1 = Zq(b[i]+b[i+step]);
            t2 = Zq(b[i]-b[i+step])*zeta;
                    
            b[i     ] = (t1-t2)*3^(Radix3);
            b[i+step] = t2*x2*3^(Radix3);
            
        return b;
```

```python
def ntt_mul(n,q,a,b):
    Zq = IntegerModRing(q);

    c = [Zq(a[i]*b[i]) for i in range(n)];

    return c;
```

```python
def poly_mul(n,q,a,b):
    Zq = IntegerModRing(q);
    c = [0]*n;

    for i in range(n):
        for j in range(n-i):
            c[i+j  ] += Zq(a[i]*b[j]);

        for j in range(n-i,n):
            c[i+j-n] -= Zq(a[i]*b[j]);

    return c;
```
