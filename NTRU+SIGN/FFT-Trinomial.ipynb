```python
def gen_zetas(n):
        
    fac = factor(n);
    
    if len(fac) == 2:
        Radix2 = fac[0][1];
        Radix3 = fac[1][1];
    
    if len(fac) == 1:
        if fac[0][0] == 2:
            Radix2 = fac[0][1];
            Radix3 = 0;

        if fac[0][0] == 3:
            Radix2 = 0;
            Radix3 = fac[0][1];

    level = Radix2+Radix3;

    z = generator(n);
    
    tree = zero_matrix(ZZ,level+1,n);
    tree[0,0] = 3*n;

    zetas = [z^0];
                
    #Radix-2 NTT with Cyclotomic Polynomial
    tree[1,0] = tree[0,0] / 6;
    tree[1,1] = 5*tree[0,0] / 6;

    x = z^(tree[1,0]);
    zetas.append(x);
    
    print(x)
    
    #Radix-3 NTT
    for l in range(1,Radix3+1):
        for ii in range(2*3^(l-1)):
            tree[l+1,3*ii  ] = tree[l  ,  ii] / 3;
            tree[l+1,3*ii+1] = tree[l  ,  ii] / 3 + tree[0,0]/3;
            tree[l+1,3*ii+2] = tree[l  ,  ii] / 3 + 2*tree[0,0]/3;

            x = z^(tree[l+1,3*ii]);
            zetas.append(x);
            x = z^(tree[l+1,3*ii]*2);
            zetas.append(x);

    #Radix-2 NTT
    for l in range(Radix3+1,level):
        for ii in range(2*3^(Radix3)*2^(l-(Radix3+1))):
            tree[l+1,2*ii  ] = tree[l  ,  ii] / 2;
            tree[l+1,2*ii+1] = tree[l  ,  ii] / 2 + tree[0,0]/2;

            x = z^(tree[l+1,2*ii]);
            zetas.append(x);

    print(tree)
    return zetas;    
```

```python
def print_zetas(zetas):
    
    l = len(zetas);
    str1 = "static const complex_fp32_16 roots[%d] = {" % l
    print(str1)
    for ii in range(floor(l/2)-1):
        zeta1 = zetas[2*ii];
        zeta2 = zetas[2*ii+1];
        
        str = "\t";
        str += "{.real = %+6d, .imag = %+6d}, " % (round(zeta1.real()*R), round(zeta1.imag()*R))
        str += "{.real = %+6d, .imag = %+6d},"  % (round(zeta2.real()*R), round(zeta2.imag()*R))
        print(str);

    zeta1 = zetas[2*(floor(l/2)-1)];
    zeta2 = zetas[2*(floor(l/2)-1)+1];
    
    str = "\t";
    str += "{.real = %+6d, .imag = %+6d}, " % (round(zeta1.real()*R), round(zeta1.imag()*R))
    str += "{.real = %+6d, .imag = %+6d}"  % (round(zeta2.real()*R), round(zeta2.imag()*R))
    print(str);
    print("};");
```
