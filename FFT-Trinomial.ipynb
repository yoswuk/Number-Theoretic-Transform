```python
def gen_zetas(n):
        
    fac = factor(n);
    
    if len(fac) == 2:
        Radix2 = fac[0][1];
        Radix3 = fac[1][1];
    
    if len(fac) == 1:
        if fac[0][0] == 2:
            Radix2 = fac[0][1];
            Radix3 = 0;

        if fac[0][0] == 3:
            Radix2 = 0;
            Radix3 = fac[0][1];

    level = Radix2+Radix3;

    z = generator(n);
    
    tree = zero_matrix(ZZ,level+1,n);
    tree[0,0] = 3*n;

    zetas = [];
                
    #Radix-2 NTT with Cyclotomic Polynomial
    tree[1,0] = tree[0,0] / 6;
    tree[1,1] = 5*tree[0,0] / 6;

    x = z^(tree[1,0]);
    zetas.append(x);
    
    #Radix-3 NTT
    for l in range(1,Radix3+1):
        for i in range(2*3^(l-1)):
            tree[l+1,3*i  ] = tree[l  ,  i] / 3;
            tree[l+1,3*i+1] = tree[l  ,  i] / 3 + tree[0,0]/3;
            tree[l+1,3*i+2] = tree[l  ,  i] / 3 + 2*tree[0,0]/3;

            x = z^(tree[l+1,3*i]);
            zetas.append(x);
            x = z^(tree[l+1,3*i]*2);
            zetas.append(x);

    #Radix-2 NTT
    for l in range(Radix3+1,level):
        for i in range(2*3^(Radix3)*2^(l-(Radix3+1))):
            tree[l+1,2*i  ] = tree[l  ,  i] / 2;
            tree[l+1,2*i+1] = tree[l  ,  i] / 2 + tree[0,0]/2;

            x = z^(tree[l+1,2*i]);
            zetas.append(x);

    print(tree)
    return zetas;    
```

```python
def print_zetas(zetas):
    
    l = len(zetas);
    str1 = "static const complex_fp32_16 roots[%d] = {" % l
    print(str1)
    for i in range(floor(l/2)):
        zeta1 = zetas[2*i];
        zeta2 = zetas[2*i+1];
        
        str = "\t";
        str += "{.real = %+6d, .imag = %+6d}, " % (round(zeta1.real()*63356), round(zeta1.imag()*63356))
        str += "{.real = %+6d, .imag = %+6d},"  % (round(zeta2.real()*63356), round(zeta2.imag()*63356))
        print(str);

    zeta1 = zetas[l-1];
    
    str = "\t";
    str += "{.real = %+6d, .imag = %+6d}" % (round(zeta1.real()*63356), round(zeta1.imag()*63356))
    print(str);
    print("};");
```
